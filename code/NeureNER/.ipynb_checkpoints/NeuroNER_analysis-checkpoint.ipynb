## F1_score vs epoch and the stop metric

1.
the stop metric of 
stanfordNER:
    |newest_val - previous_val | / |newestVal| < TOL(0.0001)
    we just have this stop metric in stanfordNER
    
NeuroNER :
    the last N epochs have not shown improvements(newest_val > previous_val) on the validation set, then stop
    the only thing we can tune is the patient N


2.
but the F1 against the epoch seem quit fluctuating in our data,
so this in NeuroNER stop metric don't fit very well in our case:

    1. if we have a large N, like 10 (which is default value), we can get some things like **Disease** (long green line) in the following graph, it run for more than 8 hours for 52 epochs, but it seem it already convengence when 30 epoch.
    
    2. but if we slightly decreasing the N to 7, we will get some things like **PlanetaryNebula(purple)** and ** Continent(red)**, for them, the last 7 epochs have not shown improvements, but it seem it doesn't converge.

it is OK if we just train one model and then have a very high N to make sure convergence, but we need 1235 models.
we can't finish it in time.
that is what i means in last email that i gusse the NeuroNER assume user continue train just one model. 

3.so i implement a single stop metric into its source code:
where i combine the idea from both Stanford's and NeuroNER
   if there have been N time that : |newest_val - previous_val | / |newestVal| < TOL, then we stop.
