```python
import pandas as pd
```

#### analysis the generated feature

```python
feature_file_path ='data/features-true.txt'
ls_token = []
ls_label = []
ls_features = []
ls_n_features = []

# read data
with open (feature_file_path, 'r') as r_file:
    lines = r_file.readlines()
for line in lines:
    temp1, temp2 = line.split('\t')
    token,_,label = temp1.split(' ')
    features = temp2[4:].split('|')
    n_features = len(features)
    ls_token.append(token)
    ls_label.append(label)
    ls_features.append(features)
    ls_n_features.append(n_features)

test_sentence = []

# combine data
for i in range(16):
    temp =[]
    temp.append(ls_token[i])
    temp.append(ls_label[i])
    temp.append(ls_n_features[i])
    temp = temp + ls_features[i]
    test_sentence.append(temp)
    
# show data
header = ["token", 'label', 'n_features']
for i in range(39):
    header.append(str(i))
df = pd.DataFrame(test_sentence, columns = header)
df = df.transpose()    


df.iloc[:,[0,1,2,3,4,5]]
```
